```python
import pandas as pd
import numpy as np
import os
import re
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
## Path for files downloaded from Samsung Health
data_dir = "/persistentStorage/SHealth_DataSet/Shealth/"

hr_data_file_prefix = "com.samsung.health.heart_rate"

for file in os.listdir(data_dir):
    if ((hr_data_file_prefix in file) and (file.endswith("csv"))):
        print(file)
        hr_df = pd.read_csv(data_dir+file, header=1)
        break
        
hr_df.head(5)
```

```python
hr_df.sort_values(by="start_time", inplace=True, ascending=True)
hr_df
```

```python
hr_df.loc[:, "start_time"] = pd.to_datetime(hr_df["start_time"])
```

```python
base_val = dt.datetime(1970, 1, 1)
epoch_val = pd.to_datetime("2018-09-01 00:00:00")
```

```python
hr_df = hr_df[hr_df["start_time"] < epoch_val]
hr_df.sort_values(by="start_time", ascending=False, inplace=True)
hr_df_orig = hr_df.copy()
```

```python
for it, row in hr_df.head(1).iterrows():
    check_time = row["start_time"]
    
check_time.date()
```

```python
check_df = hr_df[hr_df["start_time"].dt.date == check_time.date()]
check_df.sort_values(by="start_time", ascending=True, inplace=True)
check_df[["max", "min", "heart_rate", "start_time", "end_time", "time_offset", "binning_data"]].head(50)
```
