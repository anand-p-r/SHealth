```python
import pandas as pd
import numpy as np
import os
import re
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
import json
from pprint import pprint
```

```python
## Path for files downloaded from Samsung Health
data_dir = "/persistentStorage/SHealth_DataSet/Shealth/"
sleep_json_dir    = data_dir + "jsons/com.samsung.shealth.sleep_data/"
activity_json_dir = data_dir + "jsons/com.samsung.shealth.activity.day_summary/" 

calorie_data_file_prefix = "com.samsung.shealth.calories_burned"

for file in os.listdir(data_dir):
    if ((calorie_data_file_prefix in file) and (file.endswith("csv"))):
        calorie_df = pd.read_csv(data_dir+file, header=1)
        break
        
calorie_df.head(5)
```

```python
base_val = dt.datetime(1970, 1, 1)
epoch_val = pd.to_datetime("2018-09-01 00:00:00")
ref_val = (epoch_val - base_val).total_seconds()
ref_val = ref_val * 1000
ref_val
```

```python
pd.to_datetime(1535760000000, unit="ms")
```

```python
calorie_df["day_time"].dtype
type(ref_val)
```

```python
calorie_df = calorie_df[calorie_df["day_time"] < ref_val]
calorie_df
```

```python
calorie_df.sort_values(by="day_time", ascending=False, inplace=True)
calorie_df.head(5)
```

```python
pd.to_datetime(1535673600000, unit="ms")
```

```python
calorie_df["tef_calorie"].describe()
```

```python
calorie_df["tef_calorie"].median()
```

```python
median_val = calorie_df[calorie_df["tef_calorie"] > 0]["tef_calorie"].median()
calorie_df[calorie_df["tef_calorie"] > 0]["tef_calorie"].describe()
```

```python
def assign_calorie(row):
    if (row["tef_calorie"] == 0):
        return (median_val)
    else:
        return (row["tef_calorie"])
```

```python
## Fill up the 0 values in tef with median of values greater than 0 and accordingly update the total calorie burn

calorie_df.loc[:, "tef_calorie"] = calorie_df.apply(lambda x: assign_calorie(x), axis=1)
```

```python
calorie_df.head(5)
```

```python
calorie_df["rest_calorie"].describe()
```

```python
#Rest + Tef calories Vs Active calories

fig, ax = plt.subplots(figsize=(20,20))

ax.scatter(x=calorie_df["active_calorie"],
           y=calorie_df["rest_calorie"]+calorie_df["tef_calorie"],
           s=calorie_df["active_time"]/(1000*60),
           c="orange",
           marker="o"
          )

ax.set_xlabel("Active Calories", fontsize=20)
ax.set_ylabel("Rest Calories", fontsize=20)

ax.set_ylim(bottom=1500)

plt.tick_params(axis="both", labelsize=12, color="black", direction="inout")
plt.grid(b=True, which="major", axis="both")
plt.show()
```
